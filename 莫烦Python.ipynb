### 2. while和for循环

#### 2.1 while循环
基本使用: 
```python
while condition:
    expressions 
```
`condition`为判断条件（`True`或者`False`）：
* 如果为`True`， 那么将执行`expressions`语句
* 否则跳出循环（若陷入死循环，使用`ctrl`+`c` 终止程序）
  
高级主题： 
除了常规比较操作:`大于(>)`、`小于(<)`、`大于等于(>=)`、`小于等于(<=)`、`等于(==)`、`不等于(!
=)` 
* 数字类型：整数和浮点也能进行`Bolean`数据操作，如该等于`0`或`0.0`，将会返回`False` 
* `None`类型：返回`False`
* 集合数据类型：如`list`、`tuple`、`dict`和`set`等，如果该集合作为`while`判断语句，如果集合中的**元素数量**为`0`，将会返回`False`, 否则返回`True`

````python
#打印1-10所有整数
condition = 1
#当condition=11时，比较结果为false跳出循环
while condition < 11: 
    print(condition)
    condition = condition + 1
````

````python
#数字类型
i = 3
#当i=0时，结果为0跳出循环
while i:
    print(i)
    i = i - 1
````

````python
#None类型
while None:
    print('None')
````

````python
#集合数据类型
a = [1, 2, 'a', 'b']
while a:
    print(a[-1])
    a = a[:len(a)-1]
````

### 8. Class类
#### 8.1 Class定义一个类
* `Class`定义一个类
* `Class`可以先定义自己的属性
* `Class`后面还可以跟`def`， 定义一个函数，类中的函数称为**方法**

````python
class Calculator:       #首字母要大写
    name = 'Good Calculator'  #class属性
    price = 18
    def add(self,x,y):
        print(self.name)
        print(x+y)
    def minus(self,x,y):
        print(x-y)
    def times(self,x,y):
        print(x*y)
    def divide(self,x,y):
        print(x/y)

cal = Calculator() 
cal.name #访问属性
cal.add(1,2) #调用方法
cal.divide(1,2)
````

#### 8.2 Class类 init
`__init__`是一个特殊的方法，每当根据类来创建实例时，Python会自动运行它  
`__init__`定义必须包含形参`self`，每当根据类来创建实例时，将自动传入实参`self`，每个与实例相关联的方法调用都自动传递实参`self`，它是一个指向实例本身的引用，让实例能够访问类中的属性和方法。以`self`为前缀的变量可供类中所有的方法调用，可通过类的任何实例来访问
***
注意:`__init__`中的下划线是**双下划线**

````python
class Calculator:
    name = 'good calculator'
    price = 18
    def __init__(self, name, price, height, width, weight):   
        self.name = name
        self.price = price
        self.h = height
        self.wi = width
        self.we = weight

c = Calculator('bad calculator', 10, 12, 8, 15)
print(c.name)
print(c.price)
````

````python
class Car:
    def __init__(self, name = 'Benz', price = 100, color = 'red'):
        self.name = name
        self.price = price
        self.color = color
        
    def car_print(self):
        print(f'name:{self.name},price:{self.price},color:{self.color}')

c = Car('bmw', 120)
c.car_print()
````

#### 8.3 Class类继承
一个类（子类）继承另一个类（父类）时，将自动获得另一个类（父类）的所有属性和方法

````python
#父类Car
class Car:
    def __init__(self, name, price, color):
        self.name = name
        self.price = price
        self.color = color
        
    def car_print(self):
        print(f'name:{self.name},price:{self.price},color:{self.color}')

#子类ElectricCar
class ElectricCar(Car):
    def __init__(self, name, price, color): 
        super().__init__(name, price, color) #函数super()使用父类的初始化方法来初始化子类的属性
        self.battery_size = 80 #给子类添加新属性
    
    def battery_size_print(self): #给子类添加新方法
        print('battery_size:', self.battery_size)

my_tesla = ElectricCar('tesla', 200, 'black')
my_tesla.car_print()
my_tesla.battery_size_print()
````

#### 10.2 list列表
* list修改:

* `list`添加:  
1.`a`列表最后面添加一个元素`n`：`a.append(n)`  
2.`a`列表指定`m`位置添加一个元素`n`：`a.insert(m,n)`    
* list移除：  
1.`a.remove(n)`若列表中有多个`n`，则表示移除第一个`n`

* list排序：  
1.从小到大排序:`a.sort()`  
2.从大到小排序:`a.sort(reverse=True)`
* list长度：  
1.`len(a)`
***
`List`中`sort()`方法原型:`list.sort(cmp=None, key=None, reverse=False)`  


````python
a = [1, 12, 2, 65, 2, 2]
print('a:', a)

#list添加元素
a.append(3)
print('a.append(3):', a)
a.insert(1, 2)
print('a.insert(1, 2):', a)

#list删除元素
a.remove(2)
print('a.remove(2):',a)
del a[0]
print('del_a[0]:', a)

#list索引
print('a[1:-1]:', a[1:-1])
print('a[2]:', a.index(2))
print('a.count(2):', a.count(2))

#list排序
a.sort()
print(a)
a.sort(reverse=True)
print(a)
````

#### 12.4 copy & deepcopy 浅复制 & 深复制

* `id`:一个对象的`id`值在CPython解释器里就代表它在内存中的地址
* 浅拷贝:当使用浅拷贝时，python只是拷贝了最外围的对象本身，内部的元素都只是拷贝了一个引用
* 深拷贝:对外围和内部元素都进行了拷贝对象本身，而不是对象的引用

````python
import copy

a = [1, 2, 3, [4, 5]]
b = copy.copy(a)
print(id(a) == id(b))
a[3][0] = 3
print(a == b)
````

````python
import copy

a = [1, 2, 3, [4, 5]]
b = copy.deepcopy(a)
print(id(a) == id(b))
a[3][0] = 3
print(a == b)
````

* `set`基本: `set`最主要的功能就是寻找一个句子或者一个`list`当中不同的元素
* 添加元素:定义好一个`set`之后我们还可以对其添加需要的元素, 使用`add`就能添加某个元素. 但是不是每一个东西都能添加, 比如一个列表
* 筛选操作:我们还能进行一些筛选操作, 比如对比另一个东西, 看看原来的`set`里有没有和它不同的。或者对比另一个东西，看看`set`里有没有相同的
* 清除元素:清除一个元素可以用`remove`或者`discard`,两者区别是`remove`的元素在`set`当中没有的话会报错，而`discard`不会。清除全部可以用`clear`

````python
char_list = ['a', 'b', 'c', 'c', 'd', 'd', 'd']
sentence = 'Are you OK?'

#找出list中不同的元素
print(set(char_list))
print(set(sentence))
print(set(char_list + list(sentence)))

#set后添加元素
unique_char = set(char_list)
unique_char.add('x')
print(unique_char)

#set后对比找相同或不同
unique_char = set(char_list)
print(unique_char.difference({'a', 'b', 'i'}))
print(unique_char.intersection({'a', 'b', 'i'}))

#清除元素
unique_char.remove('c')
print(unique_char)
unique_char.discard('d')
print(unique_char)
unique_char.clear()
print(unique_char)
````

#### 12.10 正则表达式
是用来匹配字符的一种工具，在一大串字符中寻找所需要的内容
* 简单的匹配：在文字中找到特定的内容
* 灵活匹配：如果需要找到潜在的多个可能性文字, 我们可以使用 [] 将可能的字符囊括进来
* 按类型匹配
* 重复匹配：让某个规律被重复使用
* 分组


````python
#简单的匹配
pattern1 = "cat"
pattern2 = "bird"
string = "dog runs to cat"
print(pattern1 in string)    
print(pattern2 in string)  
````

````python
#调用python内置模块re进行简单的匹配
import re

pattern1 = "cat"
pattern2 = "bird"
string = "dog runs to cat"
print(re.search(pattern1, string)) #如果匹配到, 返回一个 match的 object   
print(re.search(pattern2, string)) #如果没有匹配到, 返回 None 
````

````python
import re

#灵活匹配(需要查找"run"或"ran"，在 pattern 的" "前面需要加上一个r用来表示这是正则表达式，将au用中括号[]囊括)
ptn = r"r[au]n"
print(re.search(ptn, "dog runs to cat"))   

#匹配到了返回
ptn1 = r"H[ea]"
print(re.search(ptn1, "He is not Ha")) 
print(re.search(ptn1, "Ha is not He"))
````

````python
import re

#[]中还可以是以下这些的组合
print(re.search(r"r[A-Z]n", "dog runs to cat"))     # [A-Z] 表示的就是所有大写的英文字母
print(re.search(r"r[a-z]n", "dog runs to cat"))     # [a-z] 表示的就是所有小写的英文字母
print(re.search(r"r[0-9]n", "dog r2ns to cat"))     # [0-9] 表示的就是所有数字
print(re.search(r"r[0-9a-zA-Z]n", "dog runs to cat"))  # [0-9a-z] 表示可以是数字，任何小写字母，任何大写字母
````

````python
import re

#按类型匹配
# \d : 任何数字
# \D : 不是数字
print(re.search(r"r\dn", "dog runs to cat"))  
print(re.search(r"r\Dn", "dog runs to cat"))  

# \w : 任何大小写字母
# \W : 不是任何大小写字母
print(re.search(r"r\wn", "dog rUns to cat"))  
print(re.search(r"r\Wn", "dog r1ns to cat")) 

# \s : 任何 white space, 如 [\t\n\r\f\v]
# \S : 不是 white space
print(re.search(r"r\sn", "dog r\tns to cat"))  
print(re.search(r"r\Sn", "dog r\tns to cat")) 

# \b : 空白字符 (只在某个字的开头或结尾)
# \B : 空白字符 (不在某个字的开头或结尾)
print(re.search(r"\brnnn\b", "dog rnnn to cat"))  
print(re.search(r"r\Bnnn\B", "dog rnns to cat")) 

# \\ : 匹配 \
# . : 匹配任何字符 (除了 \n)
print(re.search(r"runs\\", "runs\ to me")) 
print(re.search(r"r.n", "r[ns to me")) 

# ^ : 匹配开头
# $ : 匹配结尾

print(re.search(r"^dog", "dog runs to cat")) 
print(re.search(r"cat$", "dog runs to cat"))

# ? : 前面的字符可有可无
print(re.search(r"Mon(day)?", "Monday"))  
print(re.search(r"Mon(day)?", "Mon"))  
````

````python
import re

#重复匹配
# * : 重复零次或多次
print(re.search(r"ab*", "a"))            
print(re.search(r"ab*", "abbbbbbb"))   

# + : 重复一次或多次
print(re.search(r"ab+", "ab"))
print(re.search(r"ab+","abbbbbbbbbbb"))

# {n} : 重复 n 次
print(re.search(r"ab{3}", "abbb"))

# {n,m} : 重复 n 至 m 次
print(re.search(r"ab{1,3}", "abbbbb")) #{n,m}中不能有空格
print(re.search(r"ab{2,10}", "abbbbb")) 
````

````python
import re

#使用 () 为找到的内容分组
#通过分组, 我们能轻松定位所找到的内容. 比如在 (\d+) 组里, 需要找到的是一些数字(.+) 组里, 需要找到的是除\n外所有字符

match = re.search(r"(\d+), Date: (.+)", "ID: 021523, Date: May/18/2024")
print(match.group())                   # 返回所有组里的内容
print(match.group(1))                  # 返回第一组里的内容
print(match.group(2))                  # 返回第二组里的内容

#组比较多的时候，使用?P<名字> 给每个组组定义一个名字当做索引,然后就能用这个名字找到这个组的内容
match = re.search(r"(?P<id>\d+), Date: (?P<date>.+)", "ID: 021523, Date: May/18/2024")
print(match.group('id'))                #返回第一组里的内容
print(match.group('date'))              #返回第二组里的内容
````

````python
import re

#使用 findall()找到全部的匹配项 
print(re.findall(r"r[ua]n", "run ran ren"))    

# | 是 or 的意思
print(re.findall(r"(run|ran)", "run ran ren")) # ['run', 'ran']
````

````python
import re

#使用sub()匹配上一些形式的字符串然后再替代掉这些字符串
print(re.sub(r"r[au]ns", "catches", "dog runs to cat"))    
````

````python
import re

#使用spli()完成字符串的分割
print(re.split(r"[,;./]", "a;b,c.d/e"))             
````

````python
import re
#使用compile完成正则重复使用

compiled_re = re.compile(r"r[ua]n") #先将正则 compile 进一个变量
print(compiled_re.search("dog ran to cat"))  
````
